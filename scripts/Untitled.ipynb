```python
import requests
import panel as pn
import hvplot.pandas
import pandas as pd
import numpy as np
```

```python
file_url = "https://raw.githubusercontent.com/TallData/Customer_Success_Dashboard/main/data/customer_detail_data.csv"
```

```python
response = requests.get(file_url)

if response.status_code == 200:
    # Save the file locally
    with open("customer_detail_data.csv", "wb") as file:
        file.write(response.content)
    print("File downloaded successfully.")
else:
    print("Failed to download the file.")
```
